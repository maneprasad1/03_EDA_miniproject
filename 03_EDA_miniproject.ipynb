Q2. Max, min , mean of asked amount,asked equity, asked valuation? 

```python
a = data[["pitcher_ask_amount","ask_equity","ask_valuation"]].agg([max,min,"mean"])
a
```

Q3. Max & Min asked- equity,asked-valuation and asked amount episode-wise? 

```python
a = data.groupby("episode_number")[["pitcher_ask_amount","ask_equity","ask_valuation"]].agg([max,min])
a
```
